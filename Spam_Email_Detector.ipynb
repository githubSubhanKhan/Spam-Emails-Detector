```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
file_path = "spam.csv"
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "shantanudhakadd/email-spam-detection-dataset-classification", file_path,
    pandas_kwargs={"encoding": "latin1"})
print(df.head())
```

```python
df.info()
```

```python
df_copy=df.copy()
df_copy.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis=1, inplace=True)
```

```python
df_copy.info()
```

```python
df_copy.replace('ham','not spam', inplace=True)
```

```python
df_copy.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
y=df_copy['v1']
X=df_copy['v2']
```

```python
le=LabelEncoder()
y_enc=le.fit_transform(y)
```

```python
tfidf=TfidfVectorizer()
x_vec=tfidf.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_vec, y_enc, test_size=0.25, random_state=42)
model=LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('Accuracy: ',accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
email = input("Enter email: ")
email_vec=tfidf.transform([email])
proba = model.predict_proba(email_vec)
print("Not spam probability:", proba[0][0])
print("Spam probability:", proba[0][1])
prediction=model.predict(email_vec)
email_enc=le.inverse_transform(prediction)
print('The email is classified as:', email_enc[0])
```
